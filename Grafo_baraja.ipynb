```python
import numpy as np
import random
import time
```

```python
valores=[i for i in range (1,14)]
palos=["corazon","picas","trebol","diamantes"]

```

# Construcción del algoritmo

```python
inicio=time.time()
camino=[]
while len(camino)<50 and time.time()-inicio<300:
    cartas=[[x,y] for x in valores for y in palos]
    random.shuffle(cartas)            
    camino=[cartas.pop(0)] #Elimina el primer valor a cartas y lo introduce dentro de camino
    c_val=camino[0][0] #Número de la carta
    c_pal=camino[0][1] #Palo de la carta
    while len(cartas)>0:
        i=0
        encontrado=False
        while i<len(cartas):
            if c_val==cartas[i][0] and (len(camino)<2 or camino[-2][0]!=c_val): ##Si coinciden por valor
                camino.append(cartas.pop(i))
                encontrado=True
                break
            if c_pal==cartas[i][1] and (len(camino)<2 or camino[-2][1]!=c_pal): ##Si coinciden por palo 
                camino.append(cartas.pop(i))
                encontrado=True
                break
            else:
                i+=1
        c_val=camino[-1][0]
        c_pal=camino[-1][1]
        if not encontrado: # Si recorrió todas las cartas y ninguna sirvió, paramos
            break    
print("El camino con mayor longitud partiendo de la carta inicial es de longitud máxima:",len(camino)-1)
print("\n")
print("El camino es", *camino)    
```
