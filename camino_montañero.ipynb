# **Single Source Shortest Path**

```python
def SSSP(ciudad):
    camin=[ciudad]
    camino=[]
    distancia_acumulada = {col: 0 if col == ciudad else np.inf for col in matriz.columns}
    while len(camin)<len(matriz):
        kilom=10000 ##Variable iniciadora
        for elemento in camin:
            for nueva in matriz.columns:
                if nueva not in camin:
                        kil=matriz.at[elemento,nueva]
                        if kil+distancia_acumulada[elemento]<=kilom:
                            distancia_acumulada[nueva]=kil+distancia_acumulada[elemento]
                            kilom=kil+distancia_acumulada[elemento]
                            elem=elemento
                            nuev = nueva
        camin.append(nuev)
        camino.append([elem, nuev])
            
    return camino
```

# **Camino más rápido por Backtracking**

```python
def backtracking(inicio,llegada):
    camino=SSSP(inicio)
    cami=[llegada]
    caminaco=0 #Ayuda a establecer una condición para que se creen los caminos
    kil=0
    while caminaco==0:
        c=0
        while c!=None:    
            c=None
            for elemento in camino:
                if elemento[1]==cami[-1]:
                    cami.append(elemento[0])
                    camino.remove(elemento)
                    c=elemento
                elif elemento[0]==cami[-1]:
                    cami.append(elemento[1])
                    camino.remove(elemento)
                    c=elemento
        if cami[-1]!=inicio:
            cami.pop() #Elimino la ultima ciudad del camino creado
        elif cami[-1]==inicio:
            caminaco=1
    cami.reverse()
    for i in range(len(cami)-1):
        kil+=matriz.at[cami[i],cami[i+1]]
    return cami,kil
```

# **Centrality algorithms**

**Degree Centrality**

```python
importancia={elemento:0 for elemento in matriz.columns}
for elemento in camino_inicial:
    importancia[elemento[0]]+=1
    importancia[elemento[1]]+=1
importancia_top = dict(sorted(importancia.items(), key=lambda item: item[1], reverse=True))
print("La lista de los puntos más importantes:",importancia_top)
```

**Closeness Centrality**

```python
cercania={col: 0 for col in matriz.columns}
for elemento_id in matriz.columns:
    for elemento in matriz.columns:
        cercania[elemento_id]+=backtracking(elemento_id,elemento)[1]
cercania={montana: (len(matriz.columns)-1)/cercania[montana] for montana in matriz.columns}
cercania_top=dict(sorted(cercania.items(),key=lambda item: round(item[1],4), reverse=True))
print(cercania_top)    
```

```python
def SSSP_multicamino(matriz, ciudad_origen):
    # Inicialización
    distancia_acumulada = {col: 0 if col == ciudad_origen else 10000 for col in matriz.columns}
    padres = {col: [] for col in matriz.columns}
    visitados = [ciudad_origen]
    
    # Mientras no hayamos visitado todas las ciudades alcanzables
    while len(visitados) < len(matriz.columns):
        kilom_minimo = 10000
        nodo_a_agregar = None
        
        # 1. Encontrar la distancia mínima global en esta iteración
        for nodo_dentro in visitados:
            for nodo_fuera in matriz.columns:
                if nodo_fuera not in visitados:
                    peso_arista = matriz.at[nodo_dentro, nodo_fuera]
                    if peso_arista > 0 and peso_arista != 10000:
                        dist_total = distancia_acumulada[nodo_dentro] + peso_arista
                        if dist_total < kilom_minimo:
                            kilom_minimo = dist_total
                            nodo_a_agregar = nodo_fuera

        # Si no encontramos ningún nodo más, salimos (grafo disconexo)
        if nodo_a_agregar is None:
            break

        # 2. SEGUNDO PASO: Identificar TODOS los padres que llegan con esa distancia mínima
        # Esto es lo que permite capturar los múltiples caminos (empates)
        for nodo_dentro in visitados:
            for nodo_fuera in matriz.columns:
                if nodo_fuera not in visitados:
                    peso_arista = matriz.at[nodo_dentro, nodo_fuera]
                    if peso_arista > 0 and peso_arista != 10000:
                        if distancia_acumulada[nodo_dentro] + peso_arista == kilom_minimo:
                            distancia_acumulada[nodo_fuera] = kilom_minimo
                            if nodo_dentro not in padres[nodo_fuera]:
                                padres[nodo_fuera].append(nodo_dentro)
        
        # Agregamos el nodo que encontramos a la lista de visitados
        visitados.append(nodo_a_agregar)

    return distancia_acumulada, padres

def calcular_puntos_betweenness(matriz):
    # Inicializamos el contador de puntos para cada ciudad
    puntos = {col: 0 for col in matriz.columns}
    
    for origen in matriz.columns:
        # Ejecutamos tu SSSP modificado para cada nodo origen
        distancias, padres = SSSP_multicamino(matriz, origen)
        
        for destino in matriz.columns:
            if origen == destino: continue
            
            # Para cada par (origen, destino), buscamos los nodos intermedios
            # Usamos una búsqueda hacia atrás (backtracking) desde el destino
            nodos_en_camino = set()
            stack = padres[destino][:] # Copiamos los padres directos del destino
            
            while stack:
                actual = stack.pop()
                if actual != origen:
                    nodos_en_camino.add(actual)
                    # Añadimos los padres del nodo actual para seguir subiendo hasta el origen
                    for p in padres[actual]:
                        if p not in stack:
                            stack.append(p)
            
            # Sumamos un punto a cada nodo que estuvo en algún camino más corto
            for nodo in nodos_en_camino:
                puntos[nodo] += 1
                
    return puntos

resultados = calcular_puntos_betweenness(matriz)
resultados_top=dict(sorted(resultados.items(),key=lambda item: round(item[1],4), reverse=True))
print(resultados_top)
```
