# **Algoritmo de Prim**

```python
camin=["Alicante"]
camino=[]
kilometraje=0
while len(camin)<len(matriz):
    kilom=1000 ##Variable iniciadora
    for nueva in matriz.columns:
        if nueva not in camin:
            for elemento in camin:
                kil=matriz.at[elemento,nueva]
                if kil<kilom:
                    kilom=kil
                    elem=elemento
                    nuev=nueva
    camin.append(nuev)
    camino.append([elem,nuev])
    kilometraje+=kilom
print("El camino sería:",camin,f"y por tanto haríamos {len(camin)} paradas")
print("El árbol de mínima expansión sería", *camino)
print("La longitud total de este árbol sería:", kilometraje,"km")

    

        
```

# **Single Source Shortest Path**

```python
camin=["Alicante"]
camino=[]
distancia_acumulada = {col: 0 if col == "Alicante" else np.inf for col in matriz.columns}
total=0
while len(camin)<len(matriz):
    kilom=10000 ##Variable iniciadora
    for elemento in camin:
        for nueva in matriz.columns:
            if nueva not in camin:
                    kil=matriz.at[elemento,nueva]
                    if kil+distancia_acumulada[elemento]<kilom:
                        distancia_acumulada[nueva]=kil+distancia_acumulada[elemento]
                        kilom=kil+distancia_acumulada[elemento]
                        elem=elemento
                        nuev = nueva
    camin.append(nuev)
    camino.append([elem, nuev])
for montana in matriz.columns:
     total+=distancia_acumulada[montana]
         
print("El camino sería:",camin,f"y por tanto haríamos {len(camin)} paradas")
print("El árbol centrado en Alicante sería:", *camino)
print("La distancia total de ese árbol mínimo sería de:",total,"km")
```
