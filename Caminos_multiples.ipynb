```python
import numpy as np
import pandas as pd
```

```python
data = {
    'A': [0, 2, 2, 10000],
    'B': [2, 0, 10000, 1],
    'C': [2, 10000, 0, 1],
    'D': [10000, 1, 1, 0]
}
matriz = pd.DataFrame(data, index=['A', 'B', 'C', 'D'])
matriz=matriz.replace(10000,np.inf)
matriz.columns
```

```python
def SSSP(ciudad):
    camin=[ciudad]
    camino=[]
    distancia_acumulada = {col: 0 if col == ciudad else np.inf for col in matriz.columns}
    aliados={col:0 for col in matriz.columns}
    while len(camin)<len(matriz):
        kilom=10000 ##Variable iniciadora
        for elemento in camin:
            for nueva in matriz.columns:
                if nueva not in camin:
                        kil=matriz.at[elemento,nueva]
                        if kil+distancia_acumulada[elemento]<kilom:
                            distancia_acumulada[nueva]=kil+distancia_acumulada[elemento]
                            kilom=kil+distancia_acumulada[elemento]
                            elem=elemento
                            nuev = nueva
                        if kil+distancia_acumulada[elemento]==kilom:
                             aliados[]
                            
                                 
        camin.append(nuev)
        camino.append([elem, nuev])
            
    return camino
```
